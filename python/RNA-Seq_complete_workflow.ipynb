## Tutorial 2. Tracking a RNA-Seq NGS workflow in full with Synapse
_Note: The goal of this demo to show how to keep track of a typical steps in a RNA-Seq NGS workflow and not necessarily a recommended RNA-Seq or differential expression analysis method_

```python
import os, glob, sys, shutil, tempfile, re
import synapseclient
from synapseclient import File, Project, Wiki, Folder
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
syn = synapseclient.login()
```

### 1. Download the raw data from Synapse

* When you are starting a new project you would have the raw data locally. For demo purposes we are going to download a subsampled small RNA-Seq dataset for adrenal and brain tissue generated by the _[Illumina Body Map project](https://www.ebi.ac.uk/arrayexpress/experiments/E-MTAB-513/)_

The subsampled raw dataset have been stored in a public synapse project. If you want to see more information about these files you can do _syn.onweb(brain_reads)_ or _syn.onweb(adrenal_reads)_ to open them on the web.
 

```python
brain_reads = 'syn2468554'
adrenal_reads = 'syn2468552'

#lets download the fastq data
brain_reads = syn.get(brain_reads)
adrenal_reads = syn.get(adrenal_reads)
```

###2. Map the raw reads

* For demo purposed we are using a small region of chr19 (300000-350000 Mb) of the hg19 reference genome. This could also be fetched from directly from synapse.

* You can then use mapper of your choice to map these reads. We will use [STAR](http://bioinformatics.oxfordjournals.org/content/early/2012/10/25/bioinformatics.bts635) to map the reads 

####2.1 Setting up the local environment

```python
local_analysis_dir = tempfile.mkdtemp(prefix='temp_Demo_RNA-Seq_workflow')
#the following should be replaced by the location of the STAR executable
STAR_executable = 'STAR'

#dir to store the STAR genome index for the reference genome
ref_genome_dir = local_analysis_dir + '/ref_genome/'
if not os.path.exists(ref_genome_dir): os.makedirs(ref_genome_dir)
    
#dir to store the commands to execute
bin_dir = local_analysis_dir + '/bin/'
if not os.path.exists(bin_dir): os.makedirs(bin_dir)

%cd $local_analysis_dir
```

####2.2 Creating a genome index

```python
#the reference genome
ref_genome = 'syn2468557'
ref_genome = syn.get(ref_genome)

#create a STAR genome index
star_genome_index = STAR_executable + ' --runMode genomeGenerate --genomeDir {genome_dir} --genomeFastaFiles {genome_fasta}'
star_genome_index = star_genome_index.format(genome_dir=ref_genome_dir, genome_fasta = ref_genome.path)
! $star_genome_index
!echo $star_genome_index >$bin_dir'star_index_command.sh'
```

####2.3 Map adrenal & brain tissue reads
construct a command to run the mapping this command could be replaced by any mapper of choice

```python
star_command = STAR_executable +' --runThreadN 1 --genomeDir ' + ref_genome_dir + ' --outFileNamePrefix  $1 --outSAMunmapped Within --readFilesIn $2' 
print star_command
star_command_file = bin_dir +'star_command.sh'
with open(star_command_file, 'w') as f:
    f.write(star_command)
!chmod u+x $star_command_file
```

Run the mapping commands

```python
!$star_command_file adrenal_ $adrenal_reads.path
!$star_command_file brain_ $brain_reads.path
```

### 3. Store the analysis provenance in Synapse

_Key things required_
* A synapse user account 
* A personal synapse project to store the data and/or analysis
For help refer to [Synapse Getting Started Guide](https://www.synapse.org/#!Help:GettingStarted)



```python
#create a project
project_name = 'RNA_Seq_Project for %s'  % syn.getUserProfile()['userName'] 
myProj = syn.store(Project(project_name))

#create a folder to store the bams
mapped_folder = syn.store(Folder('mapped_files', parent = myProj))
```

##### Share the mapped files to Synapse along with the provenance that generated them

```python
#Store the commandline used to generate index
index_command_file = syn.store(File(bin_dir + 'star_index_command.sh', parentId=mapped_folder.id))

#Store the commandline used to run star with brain alignment
star_command_file = syn.store(File(star_command_file, parentId=mapped_folder.id))


#Upload the the mapped files but chose synapseStore=False, this creates a reference to a locally stored filed.
brain_mapped_file = glob.glob(local_analysis_dir+'/brain*sam')[0]
brain_mapped_file = syn.store(File(brain_mapped_file, parent = mapped_folder.id, synapseStore=False), 
                              used = [brain_reads, ref_genome],
                              executed= [star_command_file, index_command_file])

adrenal_mapped_file = glob.glob(local_analysis_dir+'/adrenal*sam')[0]
adrenal_mapped_file = syn.store(File(adrenal_mapped_file, parent = mapped_folder.id, synapseStore=False), 
                                used = [adrenal_reads, ref_genome],
                                executed= [star_command_file, index_command_file])
```

### 3. Continue with counting the reads per gene
**and this time lets push the provenance as we go to synapse**

_Key things needed_
* genes/features to count against
* mapped files from adrenal and brain tissues
* [htseq](http://www-huber.embl.de/users/anders/HTSeq/doc/overview.html): software to count the reads falling in each gene


####3.1 fetch the annotation & create a folder to store the counts

```python
annotation_file = 'syn2468559' #pre-stored in synaspe 
annotation_file = syn.get(annotation_file)

#lets create a folder to keep the counts
counts_folder = syn.store(Folder('counts', parentId=myProj.id))
```

Go throught each aligned sam file and call htseq on them to extract the counts per gene

```python
counts_files = []

for sam in [adrenal_mapped_file, brain_mapped_file]:    
    count_file = sam.path.replace('.sam','_htseq-counts.txt')
    command = "htseq-count -f sam -s no -q %s %s > %s" % (sam.path, annotation_file.path, count_file)
    ! $command

    #store the command in Synapse
    with tempfile.NamedTemporaryFile(mode='w+t',prefix='htseq_command', suffix='.sh', delete=False) as fp:
        fp.write(command)
    htseq_command_file = syn.store(File(fp.name, parentId=counts_folder.id))
    #Store the count file
    count_file = syn.store(File(count_file, parentId=counts_folder.id), 
                           used=[sam, annotation_file], executed=htseq_command_file)
    
    #keeping for later use
    counts_files.append(count_file)
```

### 4. Merge the counts from all the samples
produce a single counts file filtering out genes with zero counts across all the samples

```python
def merge_htseq_counts(counts_files):
    """Takes a list of htseq count files and returns a merged dataframe of genes with more than 0 counts."""
    dfs = []
    for count_file in counts_files:
        df = pd.read_csv(count_file.path,sep='\t',header=None, skip_footer=5, 
                         names=('gene',os.path.basename(count_file.path)), 
                         index_col=['gene'])
        dfs.append(df[df.ix[:,0]>0])
    return pd.concat(dfs, axis=1)
```

```python
cleaned_counts  = merge_htseq_counts(counts_files)
cleaned_counts
```

```python
cleaned_counts.apply(lambda x: np.log2(x), axis=1).plot(rot=90, title="raw counts(log)", kind="bar")
plt.savefig('raw_read_counts.png')
```

####4.1 Store the final merged counts with provenance_

```python
outFile = local_analysis_dir + '/merged_counts.tsv'
cleaned_counts.to_csv(outFile, sep="\t", header=True, index=True )
cleaned_counts_syn = syn.store(File(outFile,parentId = counts_folder.id), used=[x.id for x in counts_files])
```

### 5. Lets also create a Wiki for our project
summarizing the analysis and show analysis provenance to our users

```python
#upload the plot with provenance
plot = syn.store(File('raw_read_counts.png',parent=myProj), used=[cleaned_counts_syn])
    
#Build the wiki content
wiki = Wiki(title='My first RNA-Seq project', owner = myProj.id)
counts_markdown = re.sub( ' +', '|', cleaned_counts.to_string()).strip()

markdown = """###Raw read counts:
    {counts}
                
###Genes with difference in counts:
   {plot}

###Complete Analysis Provenance
    {provenance}"""
    
plot_wiki = '${image?synapseId=%s&align=None&scale=100}' % plot.id
provenance_wiki = '${provenance?entityList=%s&depth=4&showExpand=false&displayHeightPx=1000}' % cleaned_counts_syn.id
markdown = markdown.format(counts =counts_markdown,
                           plot = plot_wiki,
                           provenance = provenance_wiki)
        
wiki['markdown'] = markdown
wiki = syn.store(wiki)
```

```python
syn.onweb(myProj)
```

delete the demo project and local analysis space

```python
syn.delete(myProj)
shutil.rmtree(local_analysis_dir)
```
